```python
#Note: Part A of sample generation
#This code genrates initial RMPEA samples at T=850K

import numpy as np
N=50000 #Number of samples
N0=5 
N1=7 
d=0.01
d1=0.05
d2=0.36

A0=np.zeros([1,N1])
A=np.arange(d1, d2, d,dtype=float) 
Nn=0
while Nn !=N+1:
    rndlist=[]
    while np.shape(rndlist)[0] !=N0:
        r=np.random.randint(0,N1)
        if r not in rndlist:
            rndlist.append(r)
    a=1
    b=[]
    c=A[np.random.randint(np.shape(A)[0])]
    b=np.append(b,np.round(c,2))
    a=a-c
    while np.shape(b)[0] <N0: 
        if a>=d1:
            if np.shape(b)[0]<=N0-2:
                c0=A[np.random.randint(np.shape(A)[0])]
                if np.sign(a-c0)>=0:
                    b=np.append(b,np.round(c0,2))
                    a=a-c0
            elif np.shape(b)[0] ==N0-1:
                if a<=d2-d:
                    b=np.append(b,np.round(a,2))
                    a=0
                else:
                    b=np.append(b,np.round(d2-d,2))
                    a=a-(d2-d)
                    b[np.argmin(b)]=b[np.argmin(b)]+np.round(a,2)
                    a=0
        else:
            b[np.argmin(b)]=b[np.argmin(b)]+np.round(a,2)
            a=0
            while np.shape(b)[0] !=N0:
                b=np.append(b,np.round(0,2))           
    T1=0
    for TOne in range(np.shape(b)[0]):
        T1=T1+b[TOne]
    if np.sign(1-T1)<0:
        b[np.argmax(b)]=b[np.argmax(b)]+(1-T1)
    elif np.sign(1-T1)>0:
        b[np.where(b == np.min(b[np.nonzero(b)]))]=b[np.where(b == np.min(b[np.nonzero(b)]))]+(1-T1)
    Sol=np.zeros([1,N1])
    for i in range(np.shape(rndlist)[0]):
        Sol[0][rndlist[i]]=b[i]
    p1=0
    for p in range(np.shape(Sol)[1]):
        if Sol[0][p] !=0:
            p1=p1+1        
    if p1>=3: 
        p1n=0
        for p1i in range (np.shape(Sol)[1]):
            if Sol[0][p1i]>=d2 or Sol[0][p1i]<d1 and Sol[0][p1i]>0 :
                p1n=p1n+1
        if p1n==0:
            A0=np.append(A0,Sol,axis=0)           
    A0=np.unique(A0,axis=0)
    Nn=np.shape(A0)[0]       
A0=A0[1:np.shape(A0)[0]][:]
A0f=np.ones([np.shape(A0)[0],1])
A0=np.append(A0,850*A0f,axis=1) 
A0=np.append(A0,100000*A0f,axis=1)
n=[]
for i in range(A0.shape[0]):
    for j in range(N1):
        if A0[i][j]==1:
            n=np.append(n,int(i))           
if np.shape(n)[0]==1:
    A0=np.delete(A0, int(n[0]), axis=0)
elif np.shape(n)[0]==2:
    A0=np.delete(A0, [int(n[0]),int(n[1])], axis=0)
elif np.shape(n)[0]==3:
    A0=np.delete(A0, [int(n[0]),int(n[1]),int(n[2])], axis=0)   
elif np.shape(n)[0]==4:
    A0=np.delete(A0, [int(n[0]),int(n[1]),int(n[2]),int(n[3])], axis=0)   
elif np.shape(n)[0]==5:
    A0=np.delete(A0, [int(n[0]),int(n[1]),int(n[2]),int(n[3]),int(n[3])], axis=0)  
for i in range(np.shape(A0)[0]):
    with open("Samples.txt","ab") as file:
        np.savetxt(file,A0[[i],:],fmt='%.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f')
```

```python
#Note: Part B of sample generation
#This code generates range of temperature for the samples created in part A.

import numpy as np
A=np.loadtxt('Samples.txt')[:,:]
B=np.loadtxt('Element.txt') #r VEC X Tm
E=np.zeros([0,A.shape[1]]) 
for i in range(A.shape[0]):
    if np.remainder(i,10000)==0:
        print('i=',i)
    T=0
    for j in range(A.shape[1]-2):
        T=T+A[i][j]*B[j][3]
    C=np.linspace(850,0.8*T,7)
    for k in range(1,C.shape[0]):
        D=A[[i],:A.shape[1]-2] 
        D=np.append(D,np.round(C[k])) 
        D=np.append(D,100000)
        D=np.reshape(D,(1,A.shape[1]))
        E=np.vstack([E,D])        
for i in range(np.shape(E)[0]):
    with open("Samples_T.txt","ab") as file:
        np.savetxt(file,E[[i],:],fmt='%.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f')
```
