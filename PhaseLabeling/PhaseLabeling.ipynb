```python
#Note: This code label the Samples.txt and Samples_T.txt using Thermo-calc. 

%%capture
import concurrent.futures
from matplotlib import pyplot as plt
from tc_python import *
import os
path1 = os.path.abspath("")
import time
import numpy as np


def do_step(param): 
    element=[]
    if param['ti']!=0:
        element=np.append(element,"Ti")
    if param['fe']!=0:
        element=np.append(element,"Fe")
    if param['al']!=0:
        element=np.append(element,"Al")       
    if param['v']!=0:
        element=np.append(element,"V") 
    if param['ni']!=0:
        element=np.append(element,"Ni")       
    if param['nb']!=0:
        element=np.append(element,"Nb")  
    if param['zr']!=0:
        element=np.append(element,"Zr")  

        
    try:
        with TCPython() as start:
            calculation = (start.select_database_and_elements("TCHEA6", element).
                           without_default_phases().
                           select_phase("SIGMA").
                           select_phase("C14_LAVES").
                           select_phase("C36_LAVES").
                           select_phase("C15_LAVES").
                           select_phase("FCC_A1").
                           select_phase("LIQUID").
                           select_phase("BCC_B2").
                           select_phase("HCP_A3").
                           select_phase("BCC_A2").
                           select_phase("HEUSLER_L21").
                           get_system().
                           with_property_diagram_calculation().
                           with_axis(CalculationAxis(ThermodynamicQuantity.temperature()).
                                     set_min(param['Temp']).
                                     set_max(param['Temp']+1).
                                     with_axis_type(Linear().set_min_nr_of_steps(1))))

            count = sum(1 for component in param.values() if component != 0)-1 
            numphase=0
            if param["ti"] !=0 and numphase<count-1:
                calculation.set_condition(ThermodynamicQuantity.mole_fraction_of_a_component("Ti"), param["ti"] / 1)
                numphase=numphase+1
            if param["fe"] !=0 and numphase<count-1:
                calculation.set_condition(ThermodynamicQuantity.mole_fraction_of_a_component("Fe"), param["fe"] / 1)
                numphase=numphase+1
            if param["al"] !=0 and numphase<count-1:
                calculation.set_condition(ThermodynamicQuantity.mole_fraction_of_a_component("Al"), param["al"] / 1)
                numphase=numphase+1
            if param["v"] !=0 and numphase<count-1:
                calculation.set_condition(ThermodynamicQuantity.mole_fraction_of_a_component("V"), param["v"] / 1)
                numphase=numphase+1
            if param["ni"] !=0 and numphase<count-1:
                calculation.set_condition(ThermodynamicQuantity.mole_fraction_of_a_component("Ni"), param["ni"] / 1)
                numphase=numphase+1
            if param["nb"] !=0 and numphase<count-1:
                calculation.set_condition(ThermodynamicQuantity.mole_fraction_of_a_component("Nb"), param["nb"] / 1)
                numphase=numphase+1
            if param["zr"] !=0 and numphase<count-1: 
                calculation.set_condition(ThermodynamicQuantity.mole_fraction_of_a_component("Zr"), param["zr"] / 1)
                numphase=numphase+1
      


            r = (calculation.
                 calculate().
                 set_phase_name_style(PhaseNameStyle.ORDERING_DESCRIPTION).
                 get_values_grouped_by_quantity_of(ThermodynamicQuantity.temperature(),
                                                   ThermodynamicQuantity.mole_fraction_of_a_phase(ALL_PHASES)))

    except:
        r="NAN"
        
    return r

def Batch_calculation(A): 
    parameters = [
        {"index": 0, "ti": A[0,1], "fe": A[0,2],"al": A[0,3],"v": A[0,4],"ni":A[0,5],"nb": A[0,6],"zr": A[0,7],"Temp": A[0,8]},
        {"index": 0, "ti": A[1,1], "fe": A[1,2],"al": A[1,3],"v": A[1,4],"ni":A[1,5],"nb": A[1,6],"zr": A[1,7],"Temp": A[1,8]},
        {"index": 0, "ti": A[2,1], "fe": A[2,2],"al": A[2,3],"v": A[2,4],"ni":A[2,5],"nb": A[2,6],"zr": A[2,7],"Temp": A[2,8]},
        {"index": 0, "ti": A[3,1], "fe": A[3,2],"al": A[3,3],"v": A[3,4],"ni":A[3,5],"nb": A[3,6],"zr": A[3,7],"Temp": A[3,8]},
        {"index": 0, "ti": A[4,1], "fe": A[4,2],"al": A[4,3],"v": A[4,4],"ni":A[4,5],"nb": A[4,6],"zr": A[4,7],"Temp": A[4,8]},
        {"index": 0, "ti": A[5,1], "fe": A[5,2],"al": A[5,3],"v": A[5,4],"ni":A[5,5],"nb": A[5,6],"zr": A[5,7],"Temp": A[5,8]},
        {"index": 0, "ti": A[6,1], "fe": A[6,2],"al": A[6,3],"v": A[6,4],"ni":A[6,5],"nb": A[6,6],"zr": A[6,7],"Temp": A[6,8]},
        {"index": 0, "ti": A[7,1], "fe": A[7,2],"al": A[7,3],"v": A[7,4],"ni":A[7,5],"nb": A[7,6],"zr": A[7,7],"Temp": A[7,8]},
        {"index": 0, "ti": A[8,1], "fe": A[8,2],"al": A[8,3],"v": A[8,4],"ni":A[8,5],"nb": A[8,6],"zr": A[8,7],"Temp": A[8,8]},
        {"index": 0, "ti": A[9,1], "fe": A[9,2],"al": A[9,3],"v": A[9,4],"ni":A[9,5],"nb": A[9,6],"zr": A[9,7],"Temp": A[9,8]},
        {"index": 0, "ti": A[10,1], "fe": A[10,2],"al": A[10,3],"v": A[10,4],"ni":A[10,5],"nb": A[10,6],"zr": A[10,7],"Temp": A[10,8]},
        {"index": 0, "ti": A[11,1], "fe": A[11,2],"al": A[11,3],"v": A[11,4],"ni":A[11,5],"nb": A[11,6],"zr": A[11,7],"Temp": A[11,8]},
        {"index": 0, "ti": A[12,1], "fe": A[12,2],"al": A[12,3],"v": A[12,4],"ni":A[12,5],"nb": A[12,6],"zr": A[12,7],"Temp": A[12,8]},
        {"index": 0, "ti": A[13,1], "fe": A[13,2],"al": A[13,3],"v": A[13,4],"ni":A[13,5],"nb": A[13,6],"zr": A[13,7],"Temp": A[13,8]},
        {"index": 0, "ti": A[14,1], "fe": A[14,2],"al": A[14,3],"v": A[14,4],"ni":A[14,5],"nb": A[14,6],"zr": A[14,7],"Temp": A[14,8]}
        
    ]
    return parameters

def list_stable_phases(results):
    A=np.zeros([15,8]) 
    n=0
    for index,groups in results:
        a=np.zeros([1,8]) 
        if groups!="NAN":
            for group in groups.values():
                if np.char.find(group.label, "FCC_A1", start=0, end=None) >=0:
                    a[0][0]=a[0][0]+group.y[0]
                elif np.char.find(group.label, "BCC_B2", start=0, end=None) >=0 and np.char.find(group.label, "DISORD", start=0, end=None) >=0 :
                    a[0][1]=a[0][1]+group.y[0]
                elif np.char.find(group.label, "HCP_A3", start=0, end=None) >=0 :
                    a[0][2]=a[0][2]+group.y[0]
                elif np.char.find(group.label, "BCC_B2", start=0, end=None) >=0 and np.char.find(group.label, "ORD", start=0, end=None) >=0 and np.char.find(group.label, "DISORD", start=0, end=None) <0 :
                    a[0][3]=a[0][3]+group.y[0]
                elif np.char.find(group.label, "_LAVES", start=0, end=None) >=0 :
                    a[0][4]=a[0][4]+group.y[0]
                elif np.char.find(group.label, "SIGMA", start=0, end=None) >=0 :
                    a[0][5]=a[0][5]+group.y[0] 
                elif np.char.find(group.label, "HEUSLER_L21", start=0, end=None) >=0 :
                    a[0][6]=a[0][6]+group.y[0] 
            a[0][7]=np.abs(np.round(1-(a[0][0]+a[0][1]+a[0][2]+a[0][3]+a[0][4]+a[0][5]+a[0][6]),2)) 
            A[n,:]=a
            n=n+1
        else:
            A[n,:]=99999 #To distinquish the samples that thermo-calc cannot label 
            n=n+1
    return A

def cal_phase(parameters):
    results = []
    with concurrent.futures.ProcessPoolExecutor(processes) as executor:
        for property_diagram in zip(parameters, executor.map(do_step, parameters)):
            params, calc_results = property_diagram
            results.append([params["index"], calc_results])
    return results 
 
    
C=np.loadtxt('Samples.txt')
new_column=range(0,np.shape(C)[0])
C= np.insert(C, 0, new_column, axis=1)

processes = 15 

num_batch=np.shape(C)[0]//15 

for i in range(0,num_batch): 
    A=C[i*15:(i+1)*15,0:] 
    parameters=Batch_calculation(A)
    results=cal_phase(parameters)
    B=list_stable_phases(results)
    A=np.append(A,B,axis=1)
    with open('Phases.txt', 'ab') as file:
        np.savetxt(file, A, fmt='%.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f %.2f')
```
